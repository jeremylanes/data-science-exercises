```python
# we clean price field (remove $ and covert type as float
```

```python
df_price = df.copy()
df_price.price_paid = df_price.price_paid.apply(lambda x: x.replace('$', ''))
df_price.price_paid = df_price.price_paid.astype(float)
```

```python
price_filter = df_price.price_paid >= 5.0
```

```python
df_price[price_filter]
```
